```python
# =====================================
# LAB 2: Control Statements
# =====================================

# -------------------------------------
# If-Else Example
# -------------------------------------
num = int(input("Enter a number: "))

if num > 0:
    print("Positive Number")
elif num < 0:
    print("Negative Number")
else:
    print("Zero")

# -------------------------------------
# Largest of 3 Numbers
# -------------------------------------
a = int(input("Enter A: "))
b = int(input("Enter B: "))
c = int(input("Enter C: "))

if a > b and a > c:
    print("Largest is:", a)
elif b > a and b > c:
    print("Largest is:", b)
else:
    print("Largest is:", c)

# -------------------------------------
# For Loop Example
# -------------------------------------
print("Numbers from 1 to 10:")
for i in range(1, 11):
    print(i)

# -------------------------------------
# While Loop Example
# -------------------------------------
n = 5
while n > 0:
    print("Countdown:", n)
    n -= 1

# -------------------------------------
# Factorial of a Number
# -------------------------------------
num = int(input("Enter a number for factorial: "))
fact = 1

for i in range(1, num + 1):
    fact *= i

print("Factorial =", fact)

# -------------------------------------
# Prime Number Check
# -------------------------------------
num = int(input("Enter a number to check prime: "))

if num > 1:
    for i in range(2, num):
        if num % i == 0:
            print(num, "is NOT prime")
            break
    else:
        print(num, "is Prime")
else:
    print(num, "is NOT prime")

# -------------------------------------
# Palindrome Check
# -------------------------------------
num = int(input("Enter a number to check palindrome: "))
temp = num
rev = 0

while num > 0:
    digit = num % 10
    rev = rev * 10 + digit
    num //= 10

if temp == rev:
    print(temp, "is Palindrome")
else:
    print(temp, "is NOT Palindrome")
```

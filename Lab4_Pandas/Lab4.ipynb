```python
# =====================================
# LAB 4: Pandas
# =====================================

import pandas as pd
import numpy as np

# -------------------------------------
# Creating a DataFrame
# -------------------------------------
data = {
    "Name": ["Asha", "Bala", "Charan", "Divya"],
    "Age": [20, 22, 19, 21],
    "Marks": [85, 90, 78, 88]
}

df = pd.DataFrame(data)
print("DataFrame:\n", df)

# -------------------------------------
# Basic Functions
# -------------------------------------
print("\nFirst 2 rows:\n", df.head(2))
print("\nLast 2 rows:\n", df.tail(2))
print("\nInfo:")
print(df.info())

print("\nSummary:\n", df.describe())

# -------------------------------------
# Adding New Column
# -------------------------------------
df["Passed"] = df["Marks"] > 80
print("\nUpdated DataFrame:\n", df)

# -------------------------------------
# Filtering Data
# -------------------------------------
print("\nStudents with Marks > 85:\n", df[df["Marks"] > 85])

# -------------------------------------
# Sorting
# -------------------------------------
print("\nSorted by Marks:\n", df.sort_values("Marks", ascending=False))

# -------------------------------------
# Handling Missing Values
# -------------------------------------
df.loc[2, "Marks"] = np.nan
print("\nAfter adding missing value:\n", df)

print("\nMissing Values Count:\n", df.isnull().sum())

df["Marks"] = df["Marks"].fillna(df["Marks"].mean())
print("\nAfter filling missing values:\n", df)

# -------------------------------------
# GroupBy Example
# -------------------------------------
data2 = {
    "Department": ["CSE", "CSE", "ECE", "ECE", "IT"],
    "Student": ["A", "B", "C", "D", "E"],
    "Marks": [80, 90, 70, 85, 88]
}

df2 = pd.DataFrame(data2)

print("\nGroup By Department (Mean Marks):\n", df2.groupby("Department")["Marks"].mean())
```

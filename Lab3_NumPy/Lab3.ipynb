```python
# =====================================
# LAB 3: NumPy
# =====================================

import numpy as np

# -------------------------------------
# Creating Arrays
# -------------------------------------
arr1 = np.array([10, 20, 30, 40, 50])
print("1D Array:", arr1)

arr2 = np.array([[1, 2, 3], [4, 5, 6]])
print("\n2D Array:\n", arr2)

# -------------------------------------
# Array Properties
# -------------------------------------
print("\nShape:", arr2.shape)
print("Size:", arr2.size)
print("Data Type:", arr2.dtype)

# -------------------------------------
# Zeros and Ones
# -------------------------------------
zeros = np.zeros((3, 3))
ones = np.ones((2, 4))

print("\nZeros Array:\n", zeros)
print("\nOnes Array:\n", ones)

# -------------------------------------
# Range Array
# -------------------------------------
range_arr = np.arange(1, 11)
print("\nRange Array:", range_arr)

# -------------------------------------
# Reshape
# -------------------------------------
reshape_arr = np.arange(1, 13).reshape(3, 4)
print("\nReshaped Array:\n", reshape_arr)

# -------------------------------------
# Indexing and Slicing
# -------------------------------------
print("\nFirst element:", arr1[0])
print("Last element:", arr1[-1])
print("Slice (1 to 4):", arr1[1:4])

# -------------------------------------
# Mathematical Operations
# -------------------------------------
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])

print("\nAddition:", a + b)
print("Subtraction:", a - b)
print("Multiplication:", a * b)
print("Division:", a / b)

# -------------------------------------
# Statistical Functions
# -------------------------------------
data = np.array([10, 20, 30, 40, 50])

print("\nMean:", np.mean(data))
print("Sum:", np.sum(data))
print("Max:", np.max(data))
print("Min:", np.min(data))
print("Standard Deviation:", np.std(data))

# -------------------------------------
# Random Numbers
# -------------------------------------
rand_arr = np.random.randint(1, 100, 10)
print("\nRandom Array:", rand_arr)

# -------------------------------------
# Matrix Multiplication
# -------------------------------------
mat1 = np.array([[1, 2], [3, 4]])
mat2 = np.array([[5, 6], [7, 8]])

result = np.dot(mat1, mat2)
print("\nMatrix Multiplication:\n", result)
```
